<p>Nous réaliseron ici un tri par insertion du maxi afin de trier une liste de nombres entiers dans l'ordre décroissant.

<p><ul><li>Exécuter la cellule ci-dessous avant de passer à la suite.</li></ul></p>

```python
# EXECUTE ME FIRST! #
from random import randint


def creer_liste_aleatoire(nb: int, mn: int, mx: int) -> list:
    """
       Fonction créant une liste de nombres entiers tirés au hasard entre deux
       valeurs mini et maxi.

       Entrées :
         - nb : nombre d'élements de la liste
         - mn : valeur minimale des éléments
         - mx : valeur maximale des éléments

       Retour :
         - une liste de nombres entiers tirés au hasard entre mn et mx

       Exemple :
         creer_liste_aleatoire(10, -5, 19)
    """
    return [randint(mn, mx) for i in range(nb)]
```
